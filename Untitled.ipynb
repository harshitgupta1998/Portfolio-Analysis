```python
from sklearn.impute import SimpleImputer
X=data[[' Large B/P ', ' Large ROE ', ' Large S/P ',
       ' Large Return Rate in the last quarter ', ' Large Market Value ',
        'Excess Return', 'Total Risk',
       'Abs. Win Rate', 'Rel. Win Rate',  'Excess Return.1',
       'Systematic Risk.1', 'Total Risk.1', 'Abs. Win Rate.1',
       'Rel. Win Rate.1']]
y=data['Annual Return']
X_train, X_test, Y_train, Y_test= train_test_split(X,y, test_size=0.20, random_state=0)
#fill_values = SimpleImputer(missing_values=0, strategy="mean")
#X_train = fill_values.fit_transform(X_train)
#X_test = fill_values.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
model=LinearRegression()
model.fit(X_train, Y_train)
val_predictions=model.predict(X_test)
print(mean_absolute_error(Y_test, val_predictions))
#predictions1 = linearcRegr.predict(X_test)
```

```python
#from sklearn import preprocessing
#from sklearn import utils
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
degree=9

polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(X_train,Y_train)
val_predictions=polyreg.predict(X_test)
print(mean_absolute_error(Y_test, val_predictions))
```

```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense

# define the keras model
model = Sequential()
model.add(Dense(14, input_dim=14, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(X_train, Y_train, epochs=1000, batch_size=20)
# evaluate the keras model
_, accuracy = model.evaluate(X_train, Y_train)
val_predictions = model.predict_classes(X_test)
```
